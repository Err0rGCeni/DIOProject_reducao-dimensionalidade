# Extração de Features e Redução de Dimensionalidade em Dados

```python
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
img = cv.imread("/content/sample_data/CyberLena.jpg")
cv2_imshow(img)
```

## Níveis de Cinza

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

### Manualmente

```python
img_gray_hand = np.zeros(img.shape)
# Para cada linha
for idx, row in enumerate(img):
  # Pegar cada elemento rgb
  for jdx, rgb in enumerate(row):
    # Calcular média
    p = np.average(rgb)
    # E atribuir
    img_gray_hand[idx, jdx] = p

cv2_imshow(img_gray_hand)
```

## Binarização

```python
ret, img_threshold = cv.threshold(img_gray, 100, 255, cv.THRESH_BINARY)
cv2_imshow(img_threshold)
```

### Manualmente

```python
img_thres_hand = np.zeros(img_gray_hand.shape)
THRESH = 100

for idx, row in enumerate(img_gray_hand):
  for jdx, pixel in enumerate(row):
    img_thres_hand[idx, jdx] = 255 if pixel[0] >= THRESH else  0

cv2_imshow(img_thres_hand)
```
